## 2.2 예를 지정한 처리 
```
try:
    예외가 발생 할 수도 있는 문장들
except 예외타입1:
    해당 예외가 발생할 경우 실행 할 문장
except 예외타입2:
    해당 예외가 발생할 경우 실행 할 문장
```    
**여러 예외처리를 할 경우 상위 예외타입은 가장 아래로 

````python
while True :
    try:
        x = int(input('100을 나눌 정수를 입력하세요>'))
        print('입력한 정수는 {}'.format(x))
        print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
        break
    except Exception:                                         #상위 예외타입이 먼저오면 2가지 예외처리를 다 
        print('원인 모를 예외가 발생했습니다. 다시 입력하세요')  
    except ZeroDivisionError: 
        print('0으로 나눌 수 없습니다. 다시 입력하세요')
    except ValueError:
        print('정수가 아닙니다. 다시 입력하세요')   

print("Done")

````

```

    명령어들
except [예외타입 [as e]]:
    해당 예외타입 시 수행할 명령어
[else:
    try블록을 다 수행 한 후 실행할 명령어 ]
finally : 
    try블록이나 except블록을 수행한 후 실행
```

````python
try:
    f = open('data/ch08.txt','r',encoding='utf-8')  #FileNotFoundError
    data= f.read()                 #UnicodeDecodeError   
except(FileNotFoundError, UnicodeDecodeError   ) as e:
    print(e)
else :     #try블록을 다 수행 한 후 실행
    print(data)
finally :
    f.close()
````

````python
#사용자 정의 예외 : Exception클래스를 상속받은 클래스
class LengthZeroError(Exception):
    '매개변수 길이가 0일때 발생하는 예외'
    def __init__(self, message):
        super().__init__(message)  #부모 클래스의 생성자에서 예외메세지를 셋팅 
````

````python
def args_print(*data): #튜플매개변수
    if len(data)==0:
        raise LengthZeroError("매개변수 길이가 0이면 예외발생(사용자정의 예외)")
    for item in data:
        print(item, end=" ")
    print("등을 입력하셨습니다")
````

# 4절. 추상클래스
- 추상클래스 : 추상메소드가 1개이상 포함된 객체를 생성할 수 없는 클래스
- 추상메소드 : 호출할 수 없는 메소드. 상속받은 클래스에서 재정의를 강요할 때  
- 파이썬에서의 추상클래스는 raise를 이용하거나, @abstractmethod를 이용하여 추상메소드를 구현

````python
class Shape:
    def __init__(self) :
        raise NotImplementedError('객체를 생설할 수 없음')
        pass
    def calc_area(self):
        raise NotImplementedError('추상메소드')
````

````python
from abc import abstractmethod, ABC
class Shape(ABC):
    @abstractmethod
    def __init__(self) :
        pass
    @abstractmethod
    def calc_area(self):
        pass 
````

````python
s=Shape()
````

````python
class Circle(Shape):

    def __init__(self,r=3):
        self.r=r
    def calc_area(self):
        import numpy as np
        return np.pi*(self.r*2)
````

````python
myCircle =Circle(5)
myCircle.calc_area()
````

````python
#연습문제 실습형2
def isit_int(float_num):
    return int(float_num) if float_num%1==0 else float_num

while True :
    try :
        num1 =float(input('첫 번째 숫자를 입력하세요>'))
        num2 =float(input('두 번째 숫자를 입력하세요>'))
        print(f'입력한 수는 {num1}와{num2}입니다')
        result=num1/num2
        num1 =isit_int(num1)     
        num2 =isit_int(num2)
        result=isit_int(result)
        print(f'{num1}을{num2}로 나누면 {num1/num2}입니다.')
        break
    except ZeroDivisionError: 
        print('0으로 나눌 수 없습니다. 다시 시도하세요')
    except ValueError:
        print('유효한 숫자가 아닙니다. 다시 시도하세요')        
````
