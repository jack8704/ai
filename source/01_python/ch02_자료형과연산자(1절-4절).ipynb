````python
def add(x, y=0):
    '''
    매개변수 2개를 받아 두자리 숫자로 만들어요
    x= 10의 자리수가 됨
    y= 1의 자리수가 됨(default 값 0)
    사용법 : add(2, 3)=>23
    '''
    return x*10+y  
````

정규표현식(regex)
```
    1. 참조 : https://ko.wikipedia.org/wiki/정규_표현식
    2. 간략문법
            \d : 숫자와 매치. [0-9] 표현
            \D : 숫자가 아닌 문자
            \w : 영문자나 숫자,[a-zA-Z0-9]
            \W : 영문자나 숫자가 아닌 문자[~!@#$%^&*()))_]
            .  : (문자)
            \. : (.)
            \s : ( )
            {n,} : n회 반복
            {2,4} : 2~4회 반복
            ? : (0번이나 1번 의미)
            + : (1번이상 반복)
            * : (0번 이상 반복)
             [] 중에 하나 or
            
         3. 정규표현식 연습장 : https://regexr.com/
             ex1. 전화번호 패턴 0\d{1,2}[\-)\.\s]\d{3,4}[\-\s\.]\d{4}
             ex2. 이메일 패턴 \w+@\w+\.\w+\.?\w+
                            \w[a-z0-9_]+@\w+\.\w+\.?\w+    <=_ 반영한 정규표현식
                             \w[a-z0-9_]+@\w+(\.\w+){1,2}
         
 #          {n,} : n회 반복   ex \d{2,} = 숫자가 2번 반복하는 숫자                                    
 #          {2,4} : 2~4회 반복  \d{2,4}    
 #          [abcd----z] => [a-zA-Z]로 표현
 #         [a-zA-Z0-9] #문자나 숫자 아무거나 1개
    
    \w+@\w+\.\w+\.?\w+
    \w+@\w+\.\w+\.?\w+
```
