## 5.3 논리연산자 : and, &, or, | (이항연산자)     not

```python
# & : 논리연산자, 비트연산자                  & > 비교연산 > and
print((10>3) & (10>5)) #논리연산 True & True => True
print(3 & 10) # 비트연산  수&수
# 3:0011
#10:1010
#& :0010 ->2를 출력
print(10>3 & 10>5) #10>2>5
```

```python
# and : 논리연산자, 거짓으로 판별되는 첫번째 항의 결과 반환
#       모든 항이 참이면 마지막항을 반환
print(True and True and 'abc')
print(0 and 5)
```

```python
# | : 논리연산자, 비트연산자                  & > 비교연산 > and
print((10>3) | (10>5)) #논리연산 True | True => True
print(3 | 10) # 비트연산  수&수
# 3:0011
#10:1010
#| :1011 ->11를 출력
print(10>3 | 10>5) #10>2>5
```

```python
# or : 논리연산자, 거짓으로 판별되는 첫번째 항의 결과 반환
#       모든 항이 거짓이면 마지막항을 반환
print(False or False or 'abc')
print(True or False or 'abc')
print(False or True or 'abc')
print(0 or 5)
```

```python
not 10 > 5
```

# 6절 문자열 다루기
- 문자열 슬라이싱변수[from:stop:step]
- len(문자변수) : 문자수 
- upper() : 대문자로
- lower() : 소문자로
- title() : 각 어절의 앞글자만 대문자로
- capitalize() : 첫문자만 대문자로
- count('찾을문자열') : '찾을문자열'이 몇개 들어 있는지 갯수 없으면 0
- find('찾을문자열') : '찾을문자열'이 왼쪽부터 시작하여 몇번째에 있는지 index 없으면 -1
- rfind('찾을문자열') : '찾을문자열'이 오른쪽부터 시작하여 몇번쨰에 있는지 index 없으면 -1
   cf. index('찾을문자열') : '찾을문자열'이 왼쪽부터 시작하여 몇번째에 있는지, 없으면 error 라 활용도 낮음
- startswith('특정문자열') : '특정문자열'로 시작하는지 여부
- endswith('특정문자열') : '특정문자열'로 끝나는지 여부

- strip() : 좌우 공백제거(trim)
- rstrip() : 오른쪽 공백제거
- lstrip() : 왼쪽 공백제거
- split(나눌문자) : 공백이나 다른 문자기준으로 나누어 리스트로 반환
- isdisit() : 문자열이 숫자 문자열인지 여부 
- islower() : 문자열이 다 소문자인지 여부
- isupper() : 문자열이 다 대문자인지 여부
- replace(oldstr,newstr) : 문자열안의 oldstr을 newstr로 변경한 값을 반환
