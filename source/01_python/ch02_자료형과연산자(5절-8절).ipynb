## 5.1 산술연산자 : +-*/ //(몫연산자), **(제곱), %(나머지)

```python
print('10**5(승)=',10**5)
print('4/2(결과 실수)=',4/2) #다른연산자는 정수+정수는 정수이나 나누기는 정수/정수 여도 실수로 나옴
print('4//2(몫)=',4//2)
print('4%2(나머지)=',4%2)
```

```python
#나누기를 제외한 산술연산은 정수끼리 결과는 정수, 정수와 실수 연산은 결과는 실수
print("10**5.0=",10**5.0)
print('4*2.0=',4*2.0)
```

```python
#quiz1. 정수를 입력받아 짝수인지 홀수 인지 여부를 출력
num=int(input('정수를 입력하세요'))
result = num%2
if result:
    temp = '홀수'
else :
    temp = '홀수'
print("입력하신 수는 {}.   2로 나눈 나머지는 {}, 그래서 {}를 입력하셨군요".format(num,result, temp))
```

```python
#quiz2. 정수를 입력받아 5의 배수인지  여부를 출력
num=int(input('정수를 입력하세요'))
result = num%5
if result :
    temp = '5의배수아님'
else :
    temp = '5의배수임'
print(f"입력하신수는 {num}로 5로 나눈 나머지는 {result}로 {temp}")
```

```python
#quiz2. 정수를 입력받아 5의 배수인지  여부를 출력
num=str(input('정수를 입력하세요'))
result = int(num[-1])
if result == 5 or result == 0 :
    print("5의배수입니다")
else :
    print("5의배수가아닙니다")
```

## 5.2 할당연산자 : =, +=, -=, *=, /=, //=, %=

```python
a = 10
print(a)
a += 5 # a=a+5
print(a)
a -= 3 # a=a-3
print(a)
a *=10
print(a)
```

## 5.3 논리연산자 : and, &, or, | (이항연산자)     not

```python
# & : 논리연산자, 비트연산자                  & > 비교연산 > and
print((10>3) & (10>5)) #논리연산 True & True => True
print(3 & 10) # 비트연산  수&수
# 3:0011
#10:1010
#& :0010 ->2를 출력
print(10>3 & 10>5) #10>2>5
```

```python
# and : 논리연산자, 거짓으로 판별되는 첫번째 항의 결과 반환
#       모든 항이 참이면 마지막항을 반환
print(True and True and 'abc')
print(0 and 5)
```

```python
# | : 논리연산자, 비트연산자                  & > 비교연산 > and
print((10>3) | (10>5)) #논리연산 True | True => True
print(3 | 10) # 비트연산  수&수
# 3:0011
#10:1010
#| :1011 ->11를 출력
print(10>3 | 10>5) #10>2>5
```

```python
# or : 논리연산자, 거짓으로 판별되는 첫번째 항의 결과 반환
#       모든 항이 거짓이면 마지막항을 반환
print(False or False or 'abc')
print(True or False or 'abc')
print(False or True or 'abc')
print(0 or 5)
```
