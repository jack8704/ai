# 2절. 생성자(`__init__()`)와 소멸자(`__del__()`)
```
생성자 : __init__(self,매개변수,...) 객체가 생성될 때 자동 실행
        생성시 필요한 코드(인스턴스 변수 초기화)를 포함
        없으면 default 생성자가 자동 생성
소멸자 : __del__(self) 객체가 소멸될(참조한 변수가 없어질)때 자동 실행
        인스턴스 객체의 레퍼런스 카운트가 0이 될때 실행
        소멸시 필요한 코드를 포함
```


````python
class Person :
    def __init__(self, name, gender): #생성자
        self.name = name      #인스턴스 변수
        self.gender = gender  #인스턴스 변수
        print('Person 생성자 호출(객체 생성 == 클래스형 변수 생성 == 인스턴스 메모리 생성)')
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
        print(f"{self.name}님은 {self.gender}입니다")
    def __str__(self):
        return f"{self.name}은 {self.gender}이죠"
````

※ class 내의 메소드(함수) 종류
- 인스턴스메소드(일반함수) : 첫번쨰 매개변수에 self를 꼭 명시. 객체를 통해서만 호출 할 수 있는 메소드 
                                                            (ex 객체변수.__str 로 호출가능)
- 클레스메소드 : 메소드 위에 @classmethod이 선언된 메소드. 첫번쨰 매개변수에 cls(clz)를 꼭 명시
- static메소드 : 메소드 위에 @staticmethod이 선언된 메소드. 디폴트 매개변수가 없음

````python
class Person :
    academy = 'MBC' #클래스변수
    def __init__(self, name='홍길동', gender = '남자'): #생성자
        self.name = name
        self.gender = gender
    def print_info(self): #인스턴스메소드 : 변수.print_info()로 호출
        print(f'인스턴스메소드(일반함수) : {self.academy}의 {self.name}님은 {self.gender}입니다')
    def change_info(self, name, gender='남자') : #인스턴스메소드 : 변수.change_info("이름","성별")
        self.name = name
        self.gender = gender
        print(f'인스턴스메소드(일반함수)를 통해 이름은{self.name}으로 성별은 {self.gender}로 변경')
    @classmethod
    def change_class_method(clz, academy): #클래스메소드 Person.change_class_method('KBS')
        #clz == Person
        clz.academy = academy # 클래스 영역의 클래스 변수 수정
        print(f'클래스메소드를 통해 클래스변수 academy를 {academy}로 수정')
    @staticmethod
    def print_static_method(): # static메소드는 클래스변수를 이용하는 경우보다 로직이 우선일 경우
        print(f'static 메소드 :{Person.academy}로 유지')
````

# 4절. 상속과 재정의

````python
class Shape:
    'Shape는 가로와 세로 멤버변수가 있고, 어쩌구 저쩌구 ....'
    def __init__(self, width, height):
        self.width = width
        self.height = height
        print('Shape 객체 생성')
    def __del__(self):
        print('Shape 객체 소멸')
    def __str__(self):
        return f"가로:{self.width}, 세로:{self.height}"
class Rectangle(Shape):
    def print_area(self):
        print(f"넓이 : {self.width * self.height}")
r = Rectangle(3,4)
print(r)
````

- override(함수의 재정의) 가능 : 부모클래스의 함수를 자식클래스에서 다시 정의. 함수이름, 매개변수 개수가 동일

````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
      #  print("{}님 {}".format(self.name, self.gender))
        print(self)
    def __str__(self):
        return "{}님 {}".format(self.name, self.gender)
    
class Student(Person):
    def __init__(self, name , gender, major):
#         self.name = name
#         self.gender = gender
        Person.__init__(self, name, gender)
        self.major = major
    def __del__(self):
        pass
    def __str__(self): #재정의 
        return "{}님 전공은 {}이고, {}입니다".format(self.name, self.major, self.gender)
        
````

## super() 이용
- super() : " 부모클래스의 "를 의미


````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
        print("{}님 {}".format(self.name, self.gender))
    def __str__(self):
        return "{}님 {}".format(self.name, self.gender)
    
class Student(Person):
    def __init__(self, name , gender, major):
#      Person.__init__(self, name, gender)
        super().__init__(name,gender)
        self.major = major
    def __del__(self):
        pass
    def print_info(self):
        super().print_info()
        print('전공은 {}입니다'.format(self.major))
    def __str__(self): #재정의 
        return super().__str__()+ "이고, 전공은 {}입니다".format(self.major)
````

## static 변수(정적변수) : 여러 객체들 사이에 데이터를 공유하고 싶을때 사용
* static 변수 앞에 underscore가 2개(__)를 붙이면, 클래스명._클래스명__static변수명이름으로 참조
* 클래스변수는 클래스나 인스턴스나 외부에서 바로 접근 가능. 반면 __로 시작하는 변수는 외부에서 불편하게 
  클래스명._클래스명__static변수명과 같이 접근해야 하므로 외부에서 실수로 접근하거나 변경하는 것을 방지
* static 변수는 클래스 내부에서만 접근하도록 하려는 의도

````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
        print("{}님 {}".format(self.name, self.gender))
    def __str__(self):
        return "{}님 {}".format(self.name, self.gender)
    
class Student(Person):
    __count = 0 # static변수(정적변수, 공유변수) Student형변수 몇개 생성되었는지 저장
    def __init__(self, name , gender, major):
        super().__init__(name,gender)
        self.major = major
        Student._Student__count += 1 #객체가 생성될 때마다__count1 증가
    def __del__(self):
        Student._Student__count -= 1 #객체가 소멸될 때마다__count1 감소
    def print_info(self):
        super().print_info()
        print('전공은 {}입니다-이상 {}개 생성'.format(self.major, Student._Student__count))
    def __str__(self): #재정의 
        return super().__str__()+ "이고, 전공은 {}입니다-이상 {}개 생성".format(self.major, Student._Student__count)
    @classmethod
    def get_count(cls):
        return cls.__count
````

````python
s1=Student("홍","남자","AI")
print(s1)
print(Student.get_count())
print(s1.get_count())
````

# 5절. 연습문제

````python
#연습문제1
class Shape:
    def __init__(self, x=0, y=0):
        self.x=x
        self.y=y
    def move(self,x,y):
        self.x=x
        self.y=y        
    def __str__(self):
        return "x좌표값 : {}, y좌표값 : {}".format(self.x, self.y)
    @staticmethod
    def print_static_method():
        print('static 메소드 : x,y좌표값 구현')
        
        
class Triangle(Shape) :
    __count = 0
    def __init__(self, width, height, x, y) :
        self.width = width
        self.height = height
        super().__init__(x, y)
        Triangle._Triangle__count += 1
    def __del__(self) :
        Triangle._Triangle__count -= 1     
    def calc_area(self) :
        print(f"삼각형 면적은 {self.width*self.height*0.5}입니다.")
    def __str__(self) :
        return super().__str__()+f"  삼각형 면적은 {self.width*self.height*0.5}입니다."
    @classmethod
    def get_count(cls):
        return cls.__count
````

````python
s1=Shape()
print(s1.__str__())
Shape.print_static_method()
````

````python
t1=Triangle(4,3,0,10)
t2=Triangle(3,4,5,5)
print('Triangle객체 갯수 :', Triangle.get_count())
````
