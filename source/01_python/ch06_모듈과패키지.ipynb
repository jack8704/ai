````python
#1. import 모듈명 : 모듈안의 함수, 클래스, 변수를 모듈명.함수명()으로 사용
import time #time 모듈안의 모두(함수, 변수, 클래스)를 메모리 로드
#time.__file__ #기본모듈이라 위치 정보확인시 error 발생
time.ctime() # 현재 날짜와 시간
````

````python
time.time()  #1970.1.1부터의 초
start = time.time()

cnt = 0
while cnt <= 10000000:
    cnt += 1
end = time.time()
print('반복문 1천만번 돈 시간 :',(end-start),"초")
````

````python
del time # time모듈을 메모리에서 unload
````

```
모듈/패키지이나 모듈안의 함수를 가져오는 방법
import A               - A모듈을 가져옴   
import A as B          - A모듈을 B이름으로 가져옴  
from A import B        - A모듈안의 B함수(클래스|변수)를 가져옴
from A import B as C   - A모듈안의 B함수를 C이름으로 가져옴

from A import B        - A패키지 안의 B모듈을 가져옴
from A.B import C      - A패키지 안의 B모듈 안의 C함수 가져옴
from A import B as C   - A패키지 안의 B함수를 C이름으로 가져옴
from A.B import C as D - A패키지 안의 B모듈안의 C를 D이름으로 가져옴
```
