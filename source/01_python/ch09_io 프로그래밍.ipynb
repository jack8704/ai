````python
class Member:
    def __init__(self,name,age,email,address):
        self.name =name
        self.age =age
        self.email =email
        self.address =address
    def __str__(self):
        return "{},{},{},{}".format(self.name,
                                    self.age,
                                    self.email,
                                    self.address)
    def __eq__(self, other):
        if isinstance(other, Member):
            return self.__str__() == other.__str__()
        return False
    def as_dict(self):   #객체를 딕셔너리로 return
        return {'name':self.name, 'age':self.age, 'email':self.email, 'address':self.address}
    def as_list(self): #객체를 리스트로 return
        return [self.name, self.age, self.email, self.address]
        
````

````python
#2. 객체리스트를 json파일로 저장
class Member :
    def __init__(self, name, age, email, address):
        self.name = name
        self.age = age
        self.email = email
        self.address = address
    def as_dict(self):
        '객체를 딕셔너리로 바꿔 return(json.dump함수 사용에 필요)'     
        return {'name':self.name,'age':self.age, 'email':self.email, 'address':self.address}
    def __str__(self):
        return "{}님,{},{},{}주거중".format(self.name, self.age, self.email, self.address)
    def __eq__(self,other):
        '''
        m1.__eq__(m2) 호출시 두 객체의 속성변수의 값이 모두 일치하는지 여부
        m1 == m2 구현시 m1.__eq__(m2) 호출 
        '''
        if isinstance(other,Member):
            return self.__str__() ==other.__str__()  
        else:
            return False
````

## customer 클래스 작성
    * Customer("홍길동", "010-9999-9999", "dff@hong.com , 30, 5, "vvip")
    *__init__,__str__, as_dict, to_txt_style(객체를 text파일로 출력하기 위해)

````python
#연습문제 실습형

class Customer:
    def __init__(self, name, phone, email, age, grade, etc):
        self.name = name
        self.phone = phone
        self.email = email
        self.age = age
        self.grade = grade
        self.etc = etc
    def as_dict(self):             #csv 파일로 담으려면 필요
        return self.__dict__           
    def __str__(self): #****    홍길동     010-9999-9999    k@k.com   30     vvip
        return "{:>5}\t{:3}\t{:15}\t{:15}\t{:3}\t{}".format(self.grade*"*",
                                                    self.name,
                                                    self.phone,
                                                    self.email,
                                                    self.age,
                                                    self.etc)
    def to_txt_style(self): # 홍길동     010-9999-9999    k@k.com   30     vvip
        return "{:>5}\t{:3}\t{:15}\t{:15}\t{:3}\t{}".format(self.name,
                                                   self.phone,
                                                   self.email,
                                                   self.age,
                                                   self.grade,
                                                   self.etc)

  ##  customer type만 만들었다
````

````python
def to_customer(row): # "홍길동, 010-9999-9999, k@hong.com, 30, 5, vvip\n"=>Customer객체
    data = row.strip().split(', ')
    name = data[0]
    phone = data[1]
    email = data[2]
    age   = int(data[3])
    grade = int(data[4])
    etc   = data[5]
    return Customer(name, phone, email, age, grade, etc)
````

```
    data/ch09_customers.txt 파일의 내용을 customer_list에 입력받고 customer_list를 return

    data/ch09_customers.txt 빈파일을 생성하고 customer_list=[] customer_list를 return
```

````python
def load_customers():
    customer_list = []
    #파일에 입력받은 customer 정보를 append
    try:
        with open('data/ch09_customers.txt','r', encoding='utf-8') as f :
            lines=f.readlines()
            for line in lines:
                customer = to_customer(line)
                customer_list.append(customer)
    except:
        print('데이터 파일 ch09_customers.txt가 없어서 초기화했습니다')
        with open('data/ch09_customers.txt','w', encoding='utf-8') as f : 
            f.write('')
    return customer_list
````

````python
def fn1_insert_customer_info():
    '사용자로부터 고객정보를 입력받아 Customer형 객체를 return'
    import re
    name = input('이름 : ') #2길동 abc 234 김
    name_pattern = r'[가-힣]{2,}'
    while not re.search(name_pattern,name):
        print('이름을 제대로 입력하세요(한글 2글자 이상)')
        name= input('이름 : ') 
    phone = input('전화 :')
    email = input('메일 :')
    while True:
        try:
            age = int(input('나이:'))
            if (age<=0) | (age>130):
                raise Exception("나이 범위 이상")
            break      
        except:
            print('올바른 나이를 입력하세요')
    try:
        grade = int(input('등급(1~5):'))
        if grade<1:
            print(grade=1)      
        if grade>5:
            grade=5
    except:
        grade= 1
        print('유효하지 않는 등급 입력시 1등급으로 초기화')
    etc = input('기타 정보 : ')
    return Customer(name, phone, email, age, grade, etc)
````

## 2. 전체출력


````python
def fn2_print_customers(customer_list):
    'customer_list를 출력(pdf 41p.처럼)'
    print("="*70)
    print("{:^70}".format('고객정보'))
    print("-"*70)
    print("{:>5}\t{:^3}\t{:^13}\t{:^12}\t{:3}\t{}".format('GRADE',
                                                    '이름',
                                                    '전화',
                                                    '메일',
                                                    '나이',
                                                    '기타'))
    print("-"*70)
    for customer in customer_list:
        print(customer)
````

## 3. 종료

````python
def fn3_save_customer_txt(customer_list) :
    'customer_list(객체list)의 내용을 ch09_customers.txt에 백업'
    customer_text_list=[]
    for customer in customer_list:
        text = customer.to_txt_style()+'\n'
        customer_text_list.append(text)
    with open('data/ch09_customers.txt','w',encoding='utf-8') as f:
        f.writelines(customer_text_list)
    print("백업 종료합니다. 또 뵙겠습니다")
````

````python
#3번 테스트 
fn3_save_customer_txt(customer_list)
````

````python
def main():
    global customer_list
    customer_list = load_customers() #파일내용을 customer_list로
    while True:
        print("1:입력", "2:전체출력", "3:백업후 종료",sep="|", end=" ")
        fn=int(input('메뉴선택: '))
        if fn==1:
            customer = fn1_insert_customer_info() #입력받은 내용으로 customer 객체 반환
            customer_list.append(customer)     
        elif fn ==2:
            fn2_print_customers(customer_list)
        elif fn ==3:
            fn3_save_customer_txt(customer_list)
            break
            
            
````

````python
if __name__=='__main__':
    main()
````
