```python
#연습문제1
class Shape:
    def __init__(self, x=0, y=0):
        self.x=x
        self.y=y
    def move(self,x,y):
        self.x=x
        self.y=y        
    def __str__(self):
        return "x좌표값 : {}, y좌표값 : {}".format(self.x, self.y)
    @staticmethod
    def print_static_method():
        print('static 메소드 : x,y좌표값 구현')
        
        
class Triangle(Shape) :
    __count = 0
    def __init__(self, width, height, x, y) :
        self.width = width
        self.height = height
        super().__init__(x, y)
        Triangle._Triangle__count += 1
    def __del__(self) :
        Triangle._Triangle__count -= 1     
    def calc_area(self) :
        print(f"삼각형 면적은 {self.width*self.height*0.5}입니다.")
    def __str__(self) :
        return super().__str__()+f"  삼각형 면적은 {self.width*self.height*0.5}입니다."
    @classmethod
    def get_count(cls):
        return cls.__count
```

```python
s1=Shape()
print(s1.__str__())
Shape.print_static_method()
```

```python
class Member:
    def __init__(self,name,age,email,address):
        self.name =name
        self.age =age
        self.email =email
        self.address =address
    def __str__(self):
        return "{},{},{},{}".format(self.name,
                                    self.age,
                                    self.email,
                                    self.address)
    def __eq__(self, other):
        if isinstance(other, Member):
            return self.__str__() == other.__str__()
        return False
    def as_dict(self):   #객체를 딕셔너리로 return
        return {'name':self.name, 'age':self.age, 'email':self.email, 'address':self.address}
    def as_list(self): #객체를 리스트로 return
        return [self.name, self.age, self.email, self.address]
        
```

```python
class Customer :
    def __init__(self, name, phone, email, age, grade, etc):
        self.name = name
        self.phone = phone
        self.email = email
        self.age = age
        self.grade = grade
        self.etc = etc        
    def as_dict(self):
        return self.__dict__
    def __str__(self):                                                    #화면 출력용
        return "{:>5}\t{:3}\t{:15}\t{:15}\t{:3}\t{}".format(self.grade*"*",
                                                    self.name,
                                                    self.phone,
                                                    self.email,
                                                    self.age,
                                                    self.etc)
    def to_txt_style(self):
        return "{:>5}\t{:3}\t{:15}\t{:15}\t{:3}\t{}".format(self.name,
                                                    self.phone,
                                                    self.email,
                                                    self.age,
                                                    self.grade,
                                                    self.etc)
```

```python
def to_customer(row):
    data = row.strip(),split(', ')
    name = data[0]
    phone = data[1]
    email = data[2]
    age   = int(data[3])
    grade = int(data[4])
    etc   = data[5]
    return Customer(name, phone, email, age, grade, etc)    
```

```python
def load_customer():
    customer_list=[]
    try:
        with open('data/ch09_customers.txt','r', encoding='utf-8') as f :
            lines=f.readlines
            for line in lines :
                customer = to_customer(line)
                customer_list.append(customer)
    except:
        return customer_list

```

```python
def load_customers():
    customer_list = []
    #파일에 입력받은 customer 정보를 append
    try:
        with open('data/ch09_customers.txt','r', encoding='utf-8') as f :
            lines=f.readlines()
            for line in lines:
                customer = to_customer(line)
                customer_list.append(customer)
    except:
        print('데이터 파일 ch09_customers.txt가 없어서 초기화했습니다')
        with open('data/ch09_customers.txt','w', encoding='utf-8') as f : 
            f.write('')
    return customer_list
```
