```python
def safe_index(lst,item) :
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 반환하는 함수
    item 요소가 없으면 -1을 반환
    '''
    if item in lst :
        return lst.index(item)
    return -1
```

## 1.5 원하는 요소 추출
1. 리스트 슬라이싱 : list에서 원하는 index의 데이터들을 추출(인덱스 범위를 벗어나도 에러나지 않음)
    - [from, stop, step] : from부터 stop앞까지 step씩 전진하면서 index 값 추출
     * from 생략시 처음, stop 생략시 끝, step생략시 1.   모두생략가능
     * 인덱스 음수 가능
2. <font color="red"> 리스트 컴프리핸션</font> : list에서 원하는 조건의 데이터로 리스트를 생성     
     
 ex )0번째 data 가지고와 짝수들 가지고와 5보다 큰거 가지고와 이름 3글자인거 가지고와 등등

```python
numbers = [1, 3, 5, 7, 9, 10, 12, 13, 18]
#꺼꾸로 출력, 짝수번째, 다출력
print('꺼꾸로', numbers[::-1])
print('짝수번째', numbers[1::2])
print('다출력', numbers[::])
```

```python
# 홀수들만 추출
lst =[]
for item in numbers :
    if item%2 : 
        lst.append(item)
lst
```

```python
def safe_remove(lst,remove_item):
    '''
    첫번째 매개변수 lst에서 remove_item 데이터 제거
    해당 remove_item이 lst에 없으면 메세지 출력
    '''
    if remove_item in lst :
        lst.remove(remove_item)
    else :
        print('해당 데이터는 없는 데이터입니다')
```

```python
numbers=[1, 2, 3, 4, 5]
safe_remove(numbers,5)
numbers
```

## 1.9 리스트 복제

```python
# 스칼라 변수의 복제
i= 10
copied_i=i
print(f"원본i={i}, 복사본i={copied_i} / 원본주소 : {id(i)}, 복사본주소 : {id(copied_i)}")
copied_i +=5 #copied_i =copied_i +5
print(f"원본i={i}, 복사본i={copied_i} / 원본주소 : {id(i)}, 복사본주소 : {id(copied_i)}")
```

```python
# 데이터 구조의 복제  #복제아님
numbers =[5, 1, 9]
copied_numbers = numbers  #복제 아님   할당 한거!
print(f"원본i={numbers}, 복사본i={copied_numbers} / 원본주소 : {id(numbers)}, 복사본주소 : {id(copied_numbers)}")
copied_numbers[0]=555
print(f"원본i={numbers}, 복사본i={copied_numbers} / 원본주소 : {id(numbers)}, 복사본주소 : {id(copied_numbers)}")
```

```python
#복제(=깊은 복사)
numbers =[5, 1, 9]
new_numbers = numbers.copy() #복제본(깊은복사)
print(f"원본i={numbers}, 복사본i={new_numbers} / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers.sort()
print(f"원본i={numbers}, 복사본i={new_numbers} / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```

```python
#복제(=깊은 복사)
numbers =['김', '박', '정']
new_numbers = numbers.copy() #복제본(깊은복사)
print(f"원본i={numbers}, 복사본i={new_numbers} / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers.sort(reverse=True)
print(f"원본i={numbers}, 복사본i={new_numbers} / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```
