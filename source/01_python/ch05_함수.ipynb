## 3.4 튜플 매개변수를 이용한 가변인수 설정

````python
# args앞의 * : 매개변수들(인수들)이 args 튜플로 전달
def add(*args) :
    "전달받은 매개변수들의 합을 return. 매개변수가 없을 경우 0을 return"
    tot=0
    for num in args :
        tot += num
    return tot
````

````python
print(add())
print(add(10, 2, 3))
print(add(1, 2, 3, 4, 5))
````

````python
def map_template(func,L=[]):
    result=[]
    for item in L :
        result.append(func(item))
    return result
````

## (2) filter 함수

````python
def filter_template(func, L) :
    result =[]
    for item in L :
        if func(item) :
            result.append(item)
    return result
    
````

````python
def even_chk(value) : #짝수면 True return, 홀 수면 False return
    return value%2==0

even_chk = lambda value : value%2==0
even_chk(9)
````
