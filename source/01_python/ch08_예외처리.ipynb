# 1절. 예외처리 필요성 
- 예외가 날 가능성이 있는 부분에 대해 미리 예상하고, 그에 대한 처리를 프로그래밍하는 것.
    (이유 : 안정정긍로 실행할 수 있도록 함)

```
    •파일이 없거나 쓰기 금지 로 인한 오류
    •데이터베이스 프로그래밍 시 제약조건 등에 의한 데이터베이스 서버 오류
    •네트워크 프로그래밍 시 네트워크 연결 실패 로 인한 오류
    •리스트 또는 튜플의 인덱스를 벗어난 참조 로 인한 오류
```

# 2절. try ~ except로 예외처리
## 2.1 try~except
```                                                        
                                                    #가장 간단한 예외처리 방법
try : 
    예외가 발생할 가능성이 있는 명령어들
except : 
    예외가 발생했을 경우 실행할 명령어들
```

````python
#100을 입력받은 정수값으로 나눠 출력
x = int(input('정수를 입력하세요>'))
print('입력한 정수는 {}'.format(x))
print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
#ZeroDivisionError : division by zero 예외객체 타입과 에러메세지 
#                    0을 입력했을 때 예외타입
#ValueError : invalid literal for int() with base 10: 'a'
#             invalid literal(잘못된 문자)를 입력할 경우
````

````python
#ZeroDivisionError : division by zero
#ValueError : invalid literal for int() with base 10: 'a'
try:
    x = int(input('정수를 입력하세요>'))
    print('입력한 정수는 {}'.format(x))
        print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
except:
    print('나누기에 유효한 정수가 아닙니다')
print("Done")
````

````python
#100을 나눌 유효한 정수를 입력할 때 까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈수를 출력
````

````python
while True :
    try:
        x = int(input('정수를 입력하세요>'))
        print('입력한 정수는 {}'.format(x))
        print('100을 입력한 정수로 나누면 {:.2f}'.format(100/x))
        break
    except: 
        print('나누기에 유효한 정수가 아닙니다')
    print("Done")
````
