```python
#연습문제1
class Shape:
    def __init__(self, x=0, y=0):
        self.x=x
        self.y=y
    def move(self,x,y):
        self.x=x
        self.y=y        
    def __str__(self):
        return "x좌표값 : {}, y좌표값 : {}".format(self.x, self.y)
    @staticmethod
    def print_static_method():
        print('static 메소드 : x,y좌표값 구현')
        
        
class Triangle(Shape) :
    __count = 0
    def __init__(self, width, height, x, y) :
        self.width = width
        self.height = height
        super().__init__(x, y)
        Triangle._Triangle__count += 1
    def __del__(self) :
        Triangle._Triangle__count -= 1     
    def calc_area(self) :
        print(f"삼각형 면적은 {self.width*self.height*0.5}입니다.")
    def __str__(self) :
        return super().__str__()+f"  삼각형 면적은 {self.width*self.height*0.5}입니다."
    @classmethod
    def get_count(cls):
        return cls.__count
```

```python
s1=Shape()
print(s1.__str__())
Shape.print_static_method()
```

```python
t1=Triangle(4,3,0,10)
t2=Triangle(3,4,5,5)
print('Triangle객체 갯수 :', Triangle.get_count())
```

```python
class Member:
    def __init__(self,name,age,email,address):
        self.name =name
        self.age =age
        self.email =email
        self.address =address
    def __str__(self):
        return "{},{},{},{}".format(self.name,
                                    self.age,
                                    self.email,
                                    self.address)
    def __eq__(self, other):
        if isinstance(other, Member):
            return self.__str__() == other.__str__()
        return False
    def as_dict(self):   #객체를 딕셔너리로 return
        return {'name':self.name, 'age':self.age, 'email':self.email, 'address':self.address}
    def as_list(self): #객체를 리스트로 return
        return [self.name, self.age, self.email, self.address]
        
```

```python
member_list=[]
member_dict_list=[]
with open('data/ch09_member.txt','r',encoding='utf=8') as f :
    lines =f.readlines()
for line in lines:
    #print(line, end="")
    data=line.strip().split(', ')  #.strip() \n 같은걸 빼주는 역할
    name=data[0]
    age=int(data[1])
    email=data[2]
    address=data[3]
    member=Member(name, age, email, address)    
    member_list.append(member)
    member_dict_list.append(member.__dict__)
```
