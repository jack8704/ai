# 1절 조건문(if)
```
if 조건1: 
    조건1이 참(True)일 경우 수행 할 명령어들(if문 단독 가능, 들여쓰기 중요)
elif 조건2 :
    조건1이 거짓(False)이고, 조건2가 참일 경우 수행 할 명령어들(elif는 필요한만큼 사용가능)
else : 
    모든 조건이 만족하지 않을 때 수행 할 명령어들(else는 단독사용불가)
```

# 2절. 반복문
## 2.1 for-in문
```
for 임시변수 in 나열가능한 자료 :
    반복할 문장들
else : 
    나열 가능한 자료를 다 소요하여 for블록을 다 수행한 후, 최종 한번 실행 할 문장들
    for문안의 break를 수행 할 경우, else블록을 수행하지 않음(else는 선택)
```

````python
for i in range(1, 10, 2) : #[1, 3, 5, 7, 9] 와 동일
    print(i, end="\t")
else :
    print("\nDone")
    print("End")
````

````python
for i in range(1,10) :
    if i == 5 :
        break
    print(i, end="\t")
else:
    print('\nDone')
print('End')
````

## 2.2 while문
```
while 조건 : 
    조건이 참일 경우 반복할 문장들
else :
    조건이 거짓일 경우 한번 실행할 문장(else 선택사항)
```

````python
# 1~100까지 짝수들의 합을 계산하고 출력 하시오 
i=1
even_sum=0 #누적변수
while i <= 100 :
    if i%2==0:
        even_sum += i
    i+=1 # i 변수 1증가
print('1~100까지 짝수의 합은',even_sum)
````

````python
num1=int(input("정수1 입력하세요 : "))
num2=int(input("정수2 입력하세요 : "))
odd_sum = 0
for i in range(min(num1,num2),max(num1,num2)+1):
    if i%2 :
        odd_sum += i
else :
    print(f"{num1}과{num2}사이의 홀수들의 합은 {odd_sum}입니다")
````

````python
num1=int(input("정수1 입력하세요 : "))
num2=int(input("정수2 입력하세요 : "))
odd_sum = 0
i=min(num1,num2)
while i <= max(num1, num2) :
    if i%2 :
        odd_sum += i
    i+=1 
else :
    print(f"{num1}과{num2}사이의 홀수들의 합은 {odd_sum}입니다")
````

````python
#연습문제10

#프로그램 안에서 random수가 만들어지고 그 random수 맞추기
#1단계 : 수를 못맞추면 UP/DOWN을 출력하고, 맞추면 축하메세지를 뿌리고 끝남 (1~100 사이의 수를 알아맞추기)
#2단계 : 수를 못맞추면 min~max를 조정, 맞추면 축하메세지 뿌리고 끝남
#3단계 : 2단계의 기회를 5회로 제한, 5회의 기회가 지나면 실패메세지 뿌리고 끝남


# 1단계
import random
min_= 1
max_= 100
num=  random.randint(min_,max_)
print('컴퓨터가 발생한 임의의 수 :', num)
while True : 
    guess_num = int(input("1~100 사이의 숫자를 맞춰보세요"))
    if (guess_num > 100) | (guess_num < 1) :
        print("범위에 벗어나는 숫자를 입력하셨습니다")
    elif guess_num > num :
        print("DOWN ↓")
    elif guess_num < num :
        print("UP ↑")
    else :
        print(f"축하축하 {num}를 맞추셨습니다 ")
        break
        
````

````python
#2단계
import random
min_= 1
max_= 100
num=  random.randint(min_,max_)
print('컴퓨터가 발생한 임의의 수 :', num)
while True : 
    guess_num = int(input(f"{min_}~{max_} 사이의 숫자를 맞춰보세요"))
    if (guess_num > 100) | (guess_num < 1) :
        print("범위에 벗어나는 숫자를 입력하셨습니다")
    elif guess_num > num :
        print(f"DOWN ↓") 
        max_=guess_num-1     
    elif guess_num < num :
        print("UP ↑")
        min_=guess_num+1
    else :
        print(f"축하축하 {num}를 맞추셨습니다 ")
        break
````

````python
#3단계
import random
min_= 1
max_= 100
num=  random.randint(min_,max_)
print('컴퓨터가 발생한 임의의 수 :', num)
chance=5
while chance > 0 : 
    guess_num = int(input(f"{min_}~{max_} 사이의 숫자를 맞춰보세요 (남은 기회{chance}번)"))
    if (guess_num > max_) | (guess_num < min_) :
        print("범위에 벗어나는 숫자를 입력하셨습니다")
    elif guess_num > num :
        print(f"DOWN ↓") 
        max_=guess_num-1     
    elif guess_num < num :
        print("UP ↑")
        min_=guess_num+1
    else :
        print(f"축하축하 {num}를 맞추셨습니다 ")
        break
    chance -= 1
else :
    print("기회를 모두 사용하셨습니다. 실패하셨습니다.")
    
print('Done')
````
