````python
x = np.linspace(0,9.9,10)  #x데이터는 1차원데이터
x.shape
````

````python
plt.figure(figsize=(5,5)) # 그래프 사이즈가 이번 셀만 세팅 
plt.rcParams['figure.figsize'] =(5,5) # 그래프 사이즈 셋팅(다음셀도 적용)
plt.subplot(2,1,1) #2행1열의 subplot에서 1번째 subplot을 다음과 같이 그린다
plt.scatter(x,np.sin(x))
plt.subplot(2,2,3) #2행2열의 subplot에서 3번째 subplot을 다음과 같이 그린다.
plt.plot(x,np.cos(x))
plt.subplot(2,2,4)
plt.plot(x,np.sin(x)*np.cos(x))
plt.show()
````

````python
plt.subplot(2,2,1) ; plt.scatter(x,np.sin(x)) #2행2열 1번쨰 plot
plt.subplot(2,2,2) ; plt.scatter(x,np.sin(x)) #2행2열 2번쨰 plot
plt.subplot(223) ; plt.scatter(x,np.sin(x)) #2행2열 3번쨰 plot
plt.subplot(224) ; plt.scatter(x,np.sin(x)) #2행2열 4번쨰 plot
plt.show()
````

 ### 2) subplots()함수로 서브플롯 생성   
- 처음부터 setting으로 subplots로 세팅
- plot함수에서 색상이나 스타일을 조정 : https://wikidocs.net/92085

````python
fig, axes = plt.subplots(nrows=2,ncols=2,sharex=False,sharey=False, figsize=(7,4) )
axes[0,0].plot(x,np.sin(x),c='#3ecdf3',lw=9) # color는 상기 url 참조   linewidth=9
axes[0,1].plot(x,np.cos(x), 'b:') #color='b', linestyle='--' (데시라인), : (점선)
axes[1,0].plot(x,np.tanh(x), '#6b3ef3')
axes[1,0].plot([0,2,4],[-1,0,1], 'ko-.') #color='black', marker='o', linestyle='-.'
axes[1,1].plot(x,np.sin(x)*np.cos(x), 'r--')
plt.show()
````

````python
# 1행 4열 서브플롯(for 문)
func_list=[np.sin,np.cos,np.tanh,lambda x : np.sin(x)*np.cos(x)]
fmt_list=['ro--','b--', '#00AA00', 'r:']
fig, axes= plt.subplots(1,4, figsize=(15,3), sharey=True) #y축공유

for i, ax in enumerate(axes):
    ax.plot(x,func_list[i](x), fmt_list[i])
````

## 2.3 Seaborn 함수로 그래프 그리기
### 1) 관계형 그래프
    - scatterplot, lineplot, relplot(서브플롯 제공)
    
    

````python
# 종별 다른 색상과 다른 스타일의 scatterplot
sns.scatterplot(data=iris, x='petal_length',y='petal_width',hue='species',palette='Set2',
               style='species')
plt.show()
````

````python
#종별 다른 색상과 다른 스타일의 lineplot
sns.lineplot(data=iris, x='petal_width',y='petal_length',hue='species',style='species',
            palette='Set1')
plt.show()
````

````python
#종별 scatterplot을 서브플롯으로 => relplot(관계형 그래프에서 서브플롯을 그리는 함수)
g=sns.relplot(data=iris,x='petal_length',y='petal_width',hue='species',
           kind='scatter',#kind='scatter' 기본값
           col='species', #col단위로 서브플롯 표현
           #col_wrap=2,
           #facet_kws={'sharey':False,'sharex':False} # 축공유 여부
           )
g.fig.set_size_inches(9,3)
plt.show()
````

````python
#종별 lineplot을 서브플롯으로 => relplot(관계형 그래프에서 서브플롯을 그리는 함수)
g=sns.relplot(data=iris,x='petal_length',y='petal_width',
              hue='species',
              kind='line',
              col='species', 
              )
g.fig.set_size_inches(9,3)
plt.show()
````

````python
# 서브플롯 : scatterplot, lineplot
fig, ax =plt.subplots(1,2)
sns.scatterplot(data=iris, x='petal_length',y='petal_width',hue='species', ax=ax[0])
sns.lineplot(data=iris, x='petal_length',y='petal_width',hue='species', ax=ax[1])
````
